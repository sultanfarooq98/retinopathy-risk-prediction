```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('data.csv')

# Basic information about the dataset
data.info()

# Checking for missing values
data.isnull().sum()

# Statistical summary of the data
data.describe()

# Distribution of the target variable 'Retinopathy_Level'
sns.countplot(data['Retinopathy_Level'])
plt.title('Distribution of Retinopathy Levels')
plt.show()

# Pairplot to check relationships between features and the target
sns.pairplot(data, hue='Retinopathy_Level')
plt.show()
```
